```python
import numpy as np    #importar numpy
import pandas as pd   #importar pandas
```

```python
wdata=pd.read_csv('w_data.csv',sep=';')   #cargar el archivo csv w_data
wdata.head(10)                              #mostrar las primeras lineas de w_data
```

```python
wdata.info()                   #ver info del dataset, donde hay 48843 datos con ID, y para el resto de atributos hay 48842
```

```python
pdata=pd.read_csv('p_data.csv',sep=';')    #cargar el archivo csv w_data
pdata.head(5)                               #mostrar las primeras lineas de p_data
```

```python
pdata.info()                   #ver info del dataset, donde hay 48842 datos con ID, y para el resto de atributos hay 48841
```

```python
pdata['ID'].value_counts()
```

```python
result=pd.merge(wdata, pdata, on='ID', how='outer',indicator='union')      #fusion de ambos datasets por medio del atributo en comun 'ID', llamandolo 'result'. Se incluye una columna con la union que indica que si esta en ambos datasets o no
result.head()                                                              #mostrar las primeras lineas de result o informacion fusionada de w_data y p_data
```

```python
result.info()               #la columna ID en wdata trae 48843 valores, y en pdata trae 48842, por lo tanto hay 1 valor diferente
```

```python
result['union'].value_counts()    #cuenta los valores que coinciden en ambos datasets, y el que esta solo en el dataset de la izquierda
```

##### En total hay 48843 valores de ID, y coinciden en ambas tablas 48842; mientras que 1 de los datos esta en wdata, pero no en pdata

```python
result[result['union']!='both']['ID']    #al filtrar por el nuevo atributo union, el dato que sea diferente de both es el ID 1344995623
```
